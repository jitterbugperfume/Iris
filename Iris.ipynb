I have to change the class from categorical to numeric value.
I will use LabelEncoder from pandas.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cicek = df.iloc[:,-1].values
print(cicek)
```

### Change them to numeric value.


```python
cicek = le.fit_transform(cicek)
print(cicek)
```

### Add them into that origin dataset.

```python
cicek = pd.DataFrame(data = cicek, index = range(150), columns = ['class'])
df= df.iloc[:,:4]
df = pd.concat([df,cicek], axis=1)
df.head(100)
```

## Check 0-1-2 ratio

```python
cicek_zero = len(df.loc[df['class'] == '0'])
cicek_one = len(df.loc[df['class'] == '1'])
cicek_two = len(df.loc[df['class'] == '2'])

print("Number of True Cases: {0} ({1:2.2f}%)".format(cicek_zero, (cicek_zero/ (cicek_zero + cicek_one + cicek_two )) * 100))
```
